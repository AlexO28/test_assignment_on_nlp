Цель данного ноутбука состоит в решении задачи классификации на основе catboost.

# Подготовка данных.

```python
import pandas as pd
```

```python
tab = pd.read_csv('file.csv', sep=';')
```

```python
tab.reset_index(inplace = True)
tab['index'] = tab.index
```

```python
from processing import y_dict, decode_to_target, decode_to_true_date
```

```python
tab['target'] = decode_to_target(tab)
```

```python
tab['true_date'] = decode_to_true_date(tab)
```

```python
from lemmatization import prepare_lemmatized_text, remove_punctuation
```

```python
tab['text_employer'] = remove_punctuation(tab)
```

```python
tab['text'] = tab['text_employer'].apply(lambda x: prepare_lemmatized_text(x))
```

```python
from sklearn.model_selection import train_test_split
```

```python
train_val, test = train_test_split(tab, random_state=239, test_size=0.2)
train, val = train_test_split(train_val, random_state=239, test_size=0.2)
```

```python
from processing import get_aggregated_statistics_on_words_for_category
```

```python
aggtab1 = get_aggregated_statistics_on_words_for_category(train, 'Бизнес-карта')
aggtab2 = get_aggregated_statistics_on_words_for_category(train, 'Зарплатные проекты')
aggtab3 = get_aggregated_statistics_on_words_for_category(train, 'Эквайринг')
aggtab4 = get_aggregated_statistics_on_words_for_category(train, 'Открытие банковского счета')
```

Идея состоит в том, чтобы взять по 1000 наиболее популярных слов для каждой категории и сделать маски вхождения в лемматизированные тексты, число вхождений не важно.

А потом к этим признакам, равно как и к дате уже можно применить catboost.

```python
aggtab1 = aggtab1.head(1000)
aggtab2 = aggtab2.head(1000)
aggtab3 = aggtab3.head(1000)
aggtab4 = aggtab4.head(1000)
```

```python
words = aggtab1.index.tolist()
words.extend(aggtab2.index.tolist())
words.extend(aggtab3.index.tolist())
words.extend(aggtab4.index.tolist())
```

```python
words = [x[0] for x in words]
```

```python
words = list(set(words))
```

```python
len(words)
```

```python
words = [x for x in words if len(x)>0]
```

```python
words_frame = pd.DataFrame(words, columns = ['words'])
words_frame.to_csv('words.csv')
```

```python
indices_train = train['index']
indices_val = val['index']
indices_test = test['index']
```

```python
y_train = train['ACTION_ITEM_RESULT_PRODUCT_NAME']
y_val = val['ACTION_ITEM_RESULT_PRODUCT_NAME']
y_test = test['ACTION_ITEM_RESULT_PRODUCT_NAME']
```

```python
train = train[['text', 'true_date']]
val = val[['text', 'true_date']]
test = test[['text', 'true_date']]
```

```python
from processing import add_columns_for_catboost
```

```python
train = add_columns_for_catboost(train, words)
val = add_columns_for_catboost(val, words)
test = add_columns_for_catboost(test, words)
```

# Обучение модели.

```python
from catboost import CatBoostClassifier
```

```python
del train['text']
del val['text']
del test['text']
```

```python
model = CatBoostClassifier(n_estimators=750, random_seed=239)
```

```python
y_train = y_train.apply(lambda x: y_dict[x])
y_val = y_val.apply(lambda x: y_dict[x])
y_test = y_test.apply(lambda x: y_dict[x])
```

```python
model.fit(train, y_train)
```

```python
y_pred = model.predict(train)
y_pred = [y[0] for y in y_pred]
```

```python
len(y_train[y_train == y_pred])/len(y_train)
```

```python
y_pred = model.predict(val)
y_pred = [y[0] for y in y_pred]
```

```python
len(y_val[y_val == y_pred])/len(y_val)
```

Параметр 750 подбирался на кросс-валидации.

```python
y_pred = model.predict(test)
y_pred = [y[0] for y in y_pred]
```

```python
len(y_test[y_test == y_pred])/len(y_test)
```

Как видно, catboost получил результаты сравнимые с трансформерами, но всё-таки получше.
